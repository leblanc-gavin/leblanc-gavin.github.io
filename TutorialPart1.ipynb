```python
import pandas as pd
import requests
from io import StringIO
from IPython.display import display, Markdown

# URL of the Covid-19 CSV file
csv_url = "https://api.covidtracking.com/v1/states/daily.csv"  

# Step 1: Extract
response = requests.get(csv_url)

# Read the CSV with error handling
try:
    data = pd.read_csv(StringIO(response.text), on_bad_lines='warn') 
except Exception as e:
    print(f"An error occurred: {e}")

# Step 2: Transform
def transform_data(df):
    critical_columns = [
        "date", "state", "positive", "negative", 
        "totalTestResults", "hospitalizedCurrently", 
        "hospitalizedCumulative", "inIcuCurrently", 
        "onVentilatorCurrently", "recovered", "death"
    ]
    df = df[critical_columns].copy() 
    
    # Renaming for clarity
    rename_mapping = {
        "date": "Date",
        "state": "State",
        "positive": "Positive Cases",
        "negative": "Negative Cases",
        "totalTestResults": "Total Test Results",
        "hospitalizedCurrently": "Hospitalized Currently",
        "hospitalizedCumulative": "Hospitalized Cumulative",
        "inIcuCurrently": "ICU Currently",
        "onVentilatorCurrently": "On Ventilator Currently",
        "recovered": "Recovered",
        "death": "Deaths"
    }
    df.rename(columns=rename_mapping, inplace=True)

    return df




# Transform the data
transformed_data = transform_data(data)

# Step 3: Load (Display)
display(Markdown("# States Current and Historical Data for Covid-19 Tracking"))
display(transformed_data)
```
